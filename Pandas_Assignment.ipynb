```python
#doing necessary imports
import pandas as pd
import numpy as np
```

```python
#dataframe to be cleaned
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',  
'londON_StockhOlm', 
'Budapest_PaRis', 'Brussels_londOn'], 
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085], 
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]], 
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )', 
'12. Air France', '"Swiss Air"']}) 
```

```python
df
```

```python
#Adding Values in FlightNumber by incrementing the first value of 10
for i in range(1,df['FlightNumber'].shape[0]):
    df.loc[i,'FlightNumber'] = df.loc[i-1,'FlightNumber'] + 10
```

```python
df
```

```python
#Changing the type of FlightNumber from float to int
df['FlightNumber'] = df['FlightNumber'].astype(int)
```

```python
df
```

```python
#Splitting the From and To coloumn and making new coloumns
From = []
To = []
for i in range(df['From_To'].shape[0]):
    a,b = df['From_To'][i].split('_')
    From.append(a)
    To.append(b)
df.drop('From_To',axis=1,inplace=True)
df.insert(0,'From',From)
df.insert(1,'To',To)
```

```python
df
```

```python
#Capitalizing the From coloumn
for i in range(df['From'].shape[0]):
    df.loc[i,'From'] = df.loc[i,'From'].capitalize()
    
#Capitalizing the To coloumn
for i in range(df['To'].shape[0]):
    df.loc[i,'To'] = df.loc[i,'To'].capitalize()
```

```python
df
```

```python
#To get the maximum length of list present in recent delays
maximum_len_of_list = max(map(lambda x: len(x), df.loc[:,'RecentDelays'])) 
```

```python
#Divinding the coloumn RecentDelays by its values and putting them in their own specific coloumns and 
#doing everthing dynamically nothing is static
for j in range(maximum_len_of_list):
    a = []
    for i in range(df['RecentDelays'].shape[0]):
        try:
            a.append(df.loc[i,'RecentDelays'][j])
        except IndexError:
            a.append(np.nan)
    df['delay_'+str(j+1)] = a
else:
    df.drop('RecentDelays',axis = 1,inplace = True)
```

```python
df
```

```python
#removing extra spaces, special symbols , numbers from the coloumn Airline
for i in range(df['Airline'].shape[0]):
    df.loc[i,'Airline'] = ''.join(filter(lambda s :s.isalpha() or s ==' ', df.loc[i,'Airline'])).strip()
```
